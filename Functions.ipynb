```python
import numpy as np
import matplotlib.pyplot as plt
#create input val
x=np.linspace(-5,5,1001)
#define function
y=3*x+1
#plot the graph
plt.plot(x,y,label='f(x)=3x+1')
#add labels add titles
plt.xlabel('x (domain)')
plt.ylabel('f(x) (range)')
plt.title(' graph of f(x)=3x+1')
# add grid andd legend
plt.grid()
plt.legend()
plt.show()
```

```python
#composite function
#input vals
x=np.linspace(-5,5,1000)
g=x**2
f_of_g=3*(g) +1
plt.plot(x,g,label='g(x)=x^2')
plt.plot(x,f_of_g,label='f(g(x))=3x^2+1')
plt.xlabel('x')
plt.ylabel('y')
plt.title('composite function')
plt.grid()
plt.legend()
```

```python
#input vals
x=np.linspace(-5,5,1000)
y=3*x+1
#inverse function
y_inv=(x-1)/3
plt.plot(x,y,label='f(x)=3x+1')
plt.plot(x,y_inv,label='f_inv(x)=(x-1)/3')
plt.plot(x,x,linestyle='--',label='y=x(mirror)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('inverse function')
plt.grid()
plt.legend()
```

```python
#LINEAR FUNCTION 
x=np.array([1,2,3,4,5])
y=2*x+3
plt.scatter(x,y)
plt.plot(x,y)
plt.title(" linear function")
plt.xlabel(" input")
plt.ylabel("output")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# input values
x = np.array([1,2,3,4,5]).reshape(-1,1) #-1 means cal dim itself and 1 col
y = np.array([3,4,5,6,7])

model = LinearRegression()

# train model
model.fit(x, y) #calu m and c in f=mx+c

# predict
predicted_sal = model.predict([[6]])
print("predicted sal for 6y:", predicted_sal[0])

# plot
plt.scatter(x, y, label="actual data")
plt.plot(x, model.predict(x), label="regression")
plt.xlabel("experience")
plt.ylabel("salary")
plt.title("linear regression model")
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# load dataset
url = "https://raw.githubusercontent.com/selva86/datasets/master/Advertising.csv"
data = pd.read_csv(url)

# select feature and target
x = data['TV']
y = data['sales']

# fit polynomial degree 2
coeff = np.polyfit(x, y, 2)  #find coefficient
poly = np.poly1d(coeff) #convert them into usable equation

# generate smooth curve
x_new = np.linspace(x.min(), x.max(), 100)
y_new = poly(x_new)

# plot
plt.scatter(x, y)
plt.plot(x_new, y_new)

plt.title("polynomial regression: tv budget vs sales")
plt.xlabel("tv advertising budget")
plt.ylabel("sales")

plt.show()
```

```python
#Exponential function
x=np.linspace(0,5,100)
y=np.exp(x)
plt.plot(x,y)
plt.title(" exponential function")
plt.show()
```

```python
x=np.linspace(-2,5,100)
y=2**x
y_new=0.5**x
plt.plot(x,y)
plt.plot(x,y_new)
plt.xlabel(" x val")
plt.ylabel("f(x)")
plt.title(" exponential function")
plt.grid()
plt.show()
```

```python
#Logarithmic function
#generate the val val>0
x=np.linspace(0.1,10,100)
#log function
y=np.log(x)
plt.plot(x,y)
plt.xlabel(" x val")
plt.ylabel(" log(x)")
plt.title(" logarithmic function")
plt.grid()
plt.show()
y_new=np.exp(x)
plt.plot(x,y_new,label="expo")
plt.xlabel(" x val")
plt.ylabel(" exp(x)")
plt.title(" exponential function")
plt.grid()
plt.legend()
plt.show()
```
