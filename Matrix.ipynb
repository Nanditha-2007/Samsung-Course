```python
#matrix multiplication
import numpy as np
a=np.array([[3,4,1],[0,5,2]])
b=np.array([[2,0,0],[1,0,5],[4,5,0]])
c=np.dot(a,b)
print(" matrix a:\n",a)
print("matrix b:\n",b)
print(" result:(a*b): ",c)
```

```python
#addition, scalar mul, subtraction
a=np.array([[3,4,1],[0,5,2]])
b=np.array([[2,0,0],[1,0,5]])
c=a+b
d=a-(1/2)*b
e=5*a
print("addition: ",c)
print("subtraction: ",d)
print("scalar multiplication: ",e)
```

```python
#check AB=BA
a=np.array([[2,3,1],[4,5,1]])
b=np.array([[-1,2],[4,-2],[3,6]])
print("AB=",np.dot(a,b))
print("BA=",np.dot(b,a))
```

```python
#design matrix for an linear model
#features: hour studied and sleep hrs
x=np.array([[1,2,6],[1,4,5],[1,6,7]])
#target (markx)
y=np.array([50,60,80])
#compute coefficients using normal eqn
beta=np.linalg.inv(x.T@x)@x.T@y #@ matrix mul
print(" design matrix: ",x)
print("target matrix: ",y)
print("coeffients(beta) : ",beta)
```

```python
#convert into a sparse matrix
from sklearn.feature_extraction.text import CountVectorizer #converts text(words) into numbers(matrix form)
data=["ai ml","spam data","spam ml"]
vectorizer = CountVectorizer() #identify unique words, count how many times each words repeated
X = vectorizer.fit_transform(data) #convert each word into row number
print("sparse matrix: ",X)
print("dense form: ",X.toarray())
```

```python
#check if  it is sparse matrix or not
a=np.array([[0,0,3],[0,0,0],[1,0,0]])
#count zeros
c=np.sum(a==0)
tot_ele=a.size
print("matrix: ",a)
print("zero ele: ",c)
print("tot ele: ",tot_ele)
#check sparsitiy
if c>tot_ele/2:
    print(" matrix is sparse")
else:
    print(" matrix is not a sparse")
```

```python
#convert dense matrix into a sparse matrix
from scipy.sparse import csr_matrix
import numpy as np  # Added import for numpy
a=np.array([[0,0,3],[0,0,0],[1,0,0]])
row,col=a.shape  # Changed 'matrix' to 'a' since that's the variable name
sparse=[]
for i in range(row):
    for j in range(col):
        if(a[i][j]!=0):
            # Changed to append a tuple or list containing the three values
            sparse.append((i,j,a[i][j]))  # Using tuple to store the three values
print(" sparse rep(row,col,val): ")
for item in sparse:
    print(item)
print("dense form: ",sparse.toarray())
```
