```python
import numpy as np
import matplotlib.pyplot as plt
#generate time values
x=np.linspace(0,10,100)
#sin and cos waves
y_sin=np.sin(x)
y_cos=np.cos(x)
plt.plot(x,y_sin,label="sinx")
plt.plot(x,y_cos,label="cos")

plt.title(" trigonometric graphs in data science")
plt.legend()
plt.show()
```

```python
#predict electricity demand using
#normal hour(wrong way),cyclic encoding(correct way)

import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
# load dataset
df=pd.read_csv("cyclic_feature_dataset.csv")
print("first 5 rows \n",df.head())
#features and target
#model 1: without cyclic encoding
x1=df[["hour"]]
y=df["demand"]
#model 2:with cyclic encoding
x2=df[['hour_sin','hour_cos']]
#train test split
x1_train,x1_test,y_train,y_test=train_test_split(x1,y,test_size=0.2,random_state=42)
x2_train,x2_test,_,_=train_test_split(x2,y,test_size=0.2,random_state=42)
#train models
model1=LinearRegression()
model1.fit(x1_train,y_train)
model2=LinearRegression()
model2.fit(x2_train,y_train)
#predictions
y_pred1=model1.predict(x1_test)
y_pred2=model2.predict(x2_test)
#evaluate models
mse1=mean_squared_error(y_test,y_pred1)
mse2=mean_squared_error(y_test,y_pred2)
print(" model without cyclic encoding mse",mse1)
print(" model with cyvlic encoding mse",mse2)
#visualization
plt.figure(figsize=(8,5))
plt.scatter(x1_test, y_test, label="Actual")

plt.xlabel("Hour")
plt.ylabel("Demand")
plt.legend()
plt.title("Without Cyclic Encoding")
plt.show()
plt.scatter(x1_test, y_pred1, label="Predicted (Normal Hour)")
plt.show()
```

```python
import pandas as pd
import numpy as np

# create hours (0 to 23)
hours = np.arange(0, 24)

# simulate electricity demand (simple pattern: low at night, high in evening)
demand = 50 + 30 * np.sin((2 * np.pi * hours) / 24) + np.random.normal(0, 5, 24)

# cyclic encoding
hour_sin = np.sin(2 * np.pi * hours / 24)
hour_cos = np.cos(2 * np.pi * hours / 24)

# create dataframe
df = pd.DataFrame({
    "hour": hours,
    "hour_sin": hour_sin,
    "hour_cos": hour_cos,
    "demand": demand
})

# save to CSV
df.to_csv("cyclic_feature_dataset.csv", index=False)

print("CSV file created successfully!")
print(df.head())
```

```python
#Text Simillarity
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
# real dsatset 
documents=[
    "AI for agriculter using sensors",
    "Smart farming using iot sensors",
    "Deep Learning for image recognition",
    "crop monitoring using AI and Iot"
]
#convert text to numerical vectors(TF-IDF)
vectorizer=TfidfVectorizer()
tfidf_matrix=vectorizer.fit_transform(documents)
#compute cosine similariries
cos_sim=cosine_similarity(tfidf_matrix)
#print similarity matrix
print("cosine similarity matrix:\n")
print(cos_sim)
```

```python
hours=np.array([2,4,6,8])
marks=np.array([45,55,65,80])
#compute slope using regression
slope,intercept=np.polyfit(hours,marks,1)
#angle
theta=np.degrees(np.arctan(slope))
print("slope(tan theta) : ",slope)
print(" angle: :",theta)

plt.scatter(hours,marks)
plt.plot(hours,slope*hours+intercept)
plt.title(" student performance trend")
plt.xlabel("hours studeies")
plt.ylabel("markx")
plt.show()

```
