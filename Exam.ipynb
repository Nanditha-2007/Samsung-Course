```python
import sympy as sp
x, k, a, b = sp.symbols('x k a b')  # Define all symbolic variables

fx = x**3-1
gx = sp.log(sp.log(x**2-3*k))  # Fixed: '3k' to '3*k'
hx = sp.exp(a*(x**b))
print("f(x): ", sp.diff(fx, x))
print("g(x): ", sp.diff(gx, x))
print("h(x): ", sp.diff(hx, x))
```

```python
import pandas as pd
import numpy as np

def load_boston():
    data_url="http://lib.stat.cmu.edu/datasets/boston"
    raw_df=pd.read_csv(data_url,sep="\s+",skiprows=22,header=None)
    data=np.hstack([raw_df.values[::2,:],raw_df.values[1::2,:2]])
    target=raw_df.values[1::2,2]
    return {'data':data,'target':target}

boston=load_boston()
x=boston['data']
y=boston['target']

print(" solve: ",np.linalg.lstsq(x,y))
```

```python
import sympy as sp
x, y = sp.symbols('x y')
f = sp.exp(x**2 + 2*y**2)

# Specify the variable for differentiation
print("fx: ", sp.diff(f, x))  #
print("fy: ", sp.diff(f, y))  # First derivative with respect to y
print("fxx: ", sp.diff(f, x, x))  # Second derivative with respect to x
print("fxy: ", sp.diff(f, x, y)) 
print("fyy: ", sp.diff(f, y, y))  
```
