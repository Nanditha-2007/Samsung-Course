```python
#limits
import numpy as np
import matplotlib.pyplot as plt
#loss func(error)
def loss(w):
    return (w-3)**2
#derivative (gradient)
def gradient(w):
    return 2*(w-3)
#values for curve
w_val=np.linspace(-1,6,100)
loss_val=loss(w_val)
#start point
w=0
learning_rate=0.2
#plot curve
plt.figure()
plt.plot(w_val,loss_val)
plt.title("gradient descent=learning direction ")
plt.xlabel("weight (w) ")
plt.ylabel("loss")
plt.show()
```
